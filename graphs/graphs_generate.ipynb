## Experiment 1 - Narrow vs wide query

```python
import matplotlib.pyplot as plt

labels = ["Narrow Query", "Wide Query"]
values = [150000, 1200000]

plt.figure(figsize=(7,5))

bars = plt.bar(labels, values, width=0.5)

plt.title("Experiment 1: Read Bytes", fontsize=13)
plt.xlabel("Query Type", fontsize=11)
plt.ylabel("Read Bytes", fontsize=11)

plt.ticklabel_format(style='plain', axis='y')

max_val = max(values)
plt.ylim(0, max_val * 1.25)

for bar in bars:
    h = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width()/2,
        h + max_val * 0.05, 
        f"{h:,}",            
        ha='center',
        fontsize=10
    )

plt.margins(x=0.2)

plt.tight_layout()

plt.savefig("exp1.png", dpi=130)
plt.show()
```
