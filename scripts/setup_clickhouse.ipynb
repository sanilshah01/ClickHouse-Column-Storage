## Setup ClickHouse (Local using Docker)

## 1. Install Docker
Download and install Docker Desktop:
https://www.docker.com/products/docker-desktop/

---

## 2. Run ClickHouse Container

Open terminal (PowerShell) and run:

docker run -d --name clickhouse ^
-p 8123:8123 ^
-p 9000:9000 ^
clickhouse/clickhouse-server

---

## 3. Open ClickHouse Client

docker exec -it clickhouse clickhouse-client

---

## 4. Verify Installation

Run:

SELECT version();

---

## 5. Enable Query Logs (Important for Experiments)

SYSTEM FLUSH LOGS;

---

## 6. Create Demo Table

Run the setup SQL:

(Refer to setup_demo.sql or create manually)

---

## 7. Run Experiments

Execute SQL files:

- exp1.sql
- exp2.sql
- exp3.sql
- exp4.sql
- exp5.sql
- exp6.sql

---

## Notes

- Port 8123 → HTTP interface  
- Port 9000 → Native client  
- Use SYSTEM FLUSH LOGS before reading system.query_log  

---

## Troubleshooting

### Error: "invalid reference format"
Use single line command in PowerShell:

docker run -d --name clickhouse -p 8123:8123 -p 9000:9000 clickhouse/clickhouse-server

---

### If container already exists:

docker rm -f clickhouse
